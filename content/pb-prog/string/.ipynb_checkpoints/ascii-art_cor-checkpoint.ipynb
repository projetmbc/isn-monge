Réorganiser ce qui a été déjà fait
--------------

Penser à copier-coller ce qui a déjà été tapé avant de l'adapter...

```python
# ----------------------- #
# -- FONCTIONS A CODER -- #
# ----------------------- #

def square(size):
    line = "*"*size

    for i in range(size):
        print(line)

        
def parallelogram(length, height):
    line = "*"*length

    for i in range(height):
        shift = " "*i

        print(shift + line)
    

def isotriangle(height):
    spaces = " "*(height - 1)
    line   = spaces + "*"

    for i in range(height):
        print(line)

        line = line[1:] + "**"


# ----------- #
# -- TESTS -- #
# ----------- #

print("""
-----------------------
UN CARRÉ DE DIMENSION 5
-----------------------
""")
square(size = 5)

print("""
--------------------------------------------
UN PARALLÉLOGRAMME DE HAUTEUR 3 ET LARGEUR 7
--------------------------------------------
""")
parallelogram(length = 7, height = 3)

print("""
------------------------------
TRIANGLE ISOCÈLE DE HAUTEUR 11
------------------------------
""")
isotriangle(height = 11)
```

Créer une fonction "menu" - Préversion
---------------------------------------------

Un copier coller pour penser à copier-coller ce qui a déjà été tapé avant de l'adapter...

**Attention !** Le code suivant ne gère pas encore les erreurs de saisie pour les longueurs et le symbole. Voir la version juste après.

```python
# -------------------------- #
# -- FONCTIONS GRAPHIQUES -- #
# -------------------------- #

def new_square(size, symbol):
    line = symbol*size

    for i in range(size):
        print(line)

        
def new_parallelogram(length, height, symbol):
    line = "*"*length

    for i in range(height):
        shift = " "*i

        print(shift + line)
    

def new_isotriangle(height, symbol):
    spaces = " "*(height - 1)
    line   = spaces + symbol

    for i in range(height):
        print(line)

    # Pour enlever un espace, il suffit ici de partir du 2ème caractère du texte. 
    # C'est ce que permet `ligne[1:]`.
        line = line[1:] + symbol*2

        
def new_hline(length, symbol):
    line = symbol*length
    
    print(line)

        
def new_vline(length, symbol):
    for i in range(length):   
        print(symbol)

        
# ----------------------------- #
# -- MENU POUR L'UTILISATEUR -- #
# ----------------------------- #

def premain():
# La boucle infinie tellement utile pour les inetrfaces HOMME-MACHINE.
    while True:
        print("""
------------------------------------------
Choix possible (un lettre à taper) :

    [c] Tracer un carré ASCII
    [p] Tracer un parallélogramme ASCII
    [t] Tracer un triangle isocèle ASCII
    [h] Tracer une ligne horizontale ASCII
    [v] Tracer une ligne varticale ASCII
    [q] Quitter
        """) 
        
        choice = input("Votre choix : ")

        
# On doit quitter.
        if choice == "q":
            break  # On casse la boucle juste où nou sommes.
        

# Une erreur de saisie.
        if choice not in "cpthv":
            print("\nCHOIX INCONNU ! Recommencer.")
            
            continue  # On recommence la boucle.

            
# Une figure.
        symbol = input("Symbole pour le tracé : ")
     
#  * Un carré.
        if choice == "c":
            size = int(input("Taille du carré : "))
            print()
            new_square(size, symbol)

#  * Un parallélogramme.
        elif choice == "p":
            length = int(input("Longueur du parallélogramme : "))
            height = int(input("Hauteur du parallélogramme : "))
            print()
            new_parallelogram(length, height, symbol)

#  * Un triangle isocèle.
        elif choice == "t":
            height = int(input("Hauteur du triangle : "))
            print()
            new_isotriangle(height, symbol)

#  * Une ligne horizontale.
        elif choice == "h":
            length = int(input("Longueur de la ligne horizontale : "))
            print()
            new_hline(length, symbol)

#  * Une ligne horizontale.
        elif choice == "v":
            length = int(input("Longueur de la ligne horizontale : "))
            print()
            new_vline(length, symbol)
            
    
# ------------------- #
# -- MISE EN ROUTE -- #
# ------------------- #

premain()
```

Créer une fonction "menu" - Version finale
---------------------------------------------

Le code suivant reprend le précédent en gérant en plus la récupération de dimensions correctes, et d'un symbole seul.

Comme nous manipulons des dimensions à différents endroits, il faut traiter la récupération d'une dimension dans une fonction dédiée que nous avons nommée `ask_dim`.

Nous faisons aussi une fonction `ask_symbol` afin de ne pas complexifier inutilement la fonction `main` qui est très similaire à la précédente `premain`.

```python
# ---------------------------------------- #
# -- FONCTIONS GRAPHIQUES (comme avant) -- #
# ---------------------------------------- #

def final_square(size, symbol):
    line = symbol*size

    for i in range(size):
        print(line)

        
def final_parallelogram(length, height, symbol):
    line = "*"*length

    for i in range(height):
        shift = " "*i

        print(shift + line)
    

def final_isotriangle(height, symbol):
    spaces = " "*(height - 1)
    line   = spaces + symbol

    for i in range(height):
        print(line)

    # Pour enlever un espace, il suffit ici de partir du 2ème caractère du texte. 
    # C'est ce que permet `ligne[1:]`.
        line = line[1:] + symbol*2

        
def final_hline(length, symbol):
    line = symbol*length
    
    print(line)

        
def final_vline(length, symbol):
    for i in range(length):   
        print(symbol)

        
# ---------------------------- #
# -- POUR LES VÉRIFICATIONS -- #
# ---------------------------- #

def ask_symbol():
    symbol = ""

    while len(symbol) != 1:
        symbol = input("Symbole pour le tracé : ")
    
    return symbol


def ask_dim(message):
    dim = ""
    
# Notez le moyen de tester qu'une chaîne est un entier.
    while not dim.isdigit():
        dim = input(message)
    
    dim = int(dim)

    return dim

    
# ----------------------------- #
# -- MENU POUR L'UTILISATEUR -- #
# ----------------------------- #

def main():
# La boucle infinie tellement utile pour les inetrfaces HOMME-MACHINE.
    while True:
        print("""
------------------------------------------
Choix possible (un lettre à taper) :

    [c] Tracer un carré ASCII
    [p] Tracer un parallélogramme ASCII
    [t] Tracer un triangle isocèle ASCII
    [h] Tracer une ligne horizontale ASCII
    [v] Tracer une ligne varticale ASCII
    [q] Quitter
        """) 
        
        choice = input("Votre choix : ")

        
# On doit quitter.
        if choice == "q":
            break  # On casse la boucle juste où nou sommes.
        

# Une erreur de saisie.
        if choice not in "cpthv":
            print("\nCHOIX INCONNU ! Recommencer.")
            
            continue  # On recommence la boucle.

            
# Une figure.
        symbol = ask_symbol()
     
#  * Un carré.
        if choice == "c":
            size = ask_dim(message = "Taille du carré : ")
            print()
            final_square(size, symbol)

#  * Un parallélogramme.
        elif choice == "p":
            length = ask_dim(message = "Longueur du parallélogramme : ")
            height = ask_dim(message = "Hauteur du parallélogramme : ")
            print()
            final_parallelogram(length, height, symbol)

#  * Un triangle isocèle.
        elif choice == "t":
            height = ask_dim(message = "Hauteur du triangle : ")
            print()
            final_isotriangle(height, symbol)

#  * Une ligne horizontale.
        elif choice == "h":
            length = ask_dim(message = "Longueur de la ligne horizontale : ")
            print()
            final_hline(length, symbol)

#  * Une ligne horizontale.
        elif choice == "v":
            length = ask_dim(message = "Longueur de la ligne horizontale : ")
            print()
            final_vline(length, symbol)
            
    
# ------------------- #
# -- MISE EN ROUTE -- #
# ------------------- #

main()
```
