#### Explications

Toutes les règles de mise en forme se font via l'utilisation de `style="..."`.

1. Le style des titres de niveau 1 se fait via `<h1 style="color: green ; background-color: yellow ; border-color: blue ; border-style: solid ; border-width: 5px ;">...</h1>`. Notez que l'on doit utiliser ceci à chaque nouveau titre ! Comment marchent ces commandes ?
    
    * `color: green ;` demande de mettre la couleur du texte en vert qui se dit "green" en anglais.
    
    * Sachant que "background" et "yellow" signifient "arrière-plan" et "jaune" en anglais, on comprend que `background-color: yellow ;` sert à obtenir un fond jaune.

    * Comme "border" et "blue" peuvent se traduire par "bordure" et "bleu", la commande `border-color: blue ;` permet d'obtenir une couleur bleue pour une éventuelle bordure.

    * Sans `border-style: solid ;` qui demande d'utiliser un trait plein pour la bordure, et `border-width: 5px ;` qui fixe à cinq pixels la largeur de la bordure, soit "width" en anglais, la bordure ne serait pas visible !

1. Le style des titres de niveau 2 via `<h2 style="color: red ;">...</h2>` ne demande pas plus d'explication que de savoir que "red" signifie "rouge".

1. Pour le tout premier paragraphe, nous avons utilisé `<p style="background-color: #BC5C46;">...</p>` avec une drôle de valeur pour la couleur `#BC5C46`. On utilise un format dit hexadécimal : [voir ici](http://www.rapidtables.com/web/color/color-wheel.htm) comment obtenir ce type de format facilement via une interface en ligne. 

1. Les balises `strong` dans les contenus des titres de niveau 2 utilisent toutes `style="color: yellow ; background-color: rgb(125, 125, 125) ; font-variant: small-caps ;"`. Nous avons ici deux nouveautés.

    * Pour la couleur d'arrière plan, nous utilisons un autre format de couleur à savoir le RGB pour Red-Green-Blue, soit Rouge-Vert-Bleu. [Ce lien](http://www.rapidtables.com/web/color/color-wheel.htm) donné juste avant vous permet d'obtenir ce type de format facilement via quelques clics. `rgb(125, 125, 125)` code une nuance de gris.

    * Pour comprendre `font-variant: small-caps ;` indiquons que "font" signifie "police de caractères", et que "small-caps" se traduit par "petites capitales". On demande donc ici d'utiliser des "petites" capitales pour le texte.


**IMPORTANT ! ** Il est indispensable d'utiliser des points-virgules pour séparer les différentes demandes de mise en forme.

### Mettre en forme en groupant les instructions à l'intérieur de `<head>...</head>`

Une première technique utilisée par [cette page](fine.html) est de regrouper les styles CSS dans `<head>...</head>` à l'intérieur d'une balise faite pour cela à savoir `<style>...</style>`. On constate que le code HTML dans `<body>...</body>` n'est plus pollué par des règles de mise en forme, et qu'il est aussi facile de changer une mise en forme car tout est centralisé dans `<style>...</style>`. Expliquons comment tout ceci fonctionne.

1. A l'intérieur de `h1 {...}` et `h2 {...}`, on définit les styles des titres de niveau $1$ et $2$ respectivement en suivant les mêmes règles que précédemment à l'intérieur de `style="..."`.

1. De même, `strong {...}` permet de définir le style des mots qui seront contenus dans `<strong>...</strong>`.

1. `h2 strong {...}` est un peu plus technique. Ceci permet de définir la mise en forme des mots dans `<strong>...</strong>` **à condition** que cette balise soit elle-même contenue dans `<h2>...</h2>`.

1. Si l'on sait que "first of type" signifie "le premier du type", on comprend que `p:first-of-type {...}` gère la mise en forme du tout premier paragraphe rencontré.


**AVERTISSEMENT !** Nous avons volontairement laissé de côté les styles en cascade dans notre étude de cas. Par exemple, les styles définis pour `<strong>...</strong>` vont être automatiquement appliqués à tout `<strong>...</strong>`, y compris ceux contenues dans un `<h2>...</h2>`. Donc quand on utilise `h2 strong {...}`, il y a au préalable toutes les règles définies via `strong {...}`. Les styles s'appliquent en cascade, d'où le nom "Cascading Style Sheet" pour CSS.

### Encore mieux, utiliser un fichier CSS externe

L'idéal est en fait d'utiliser un fichier CSS dédié à part du fichier HTML. On peut ainsi employer le même style pour différentes pages HTML sans trop d'efforts. [Cette page](graal.html) utilise cette technique. Son code HTML est donné ci-dessous.
Vous y verrez `<link rel="stylesheet" type="text/css" href="css_for_graal.css">` dans `<head>...</head>`.
C'est ceci qui permet d'importer les règles de styles définies dans le fichier CSS de chemin relatif `css_for_graal.css`
*(rien n'empêche d'utiliser un lien vers un fichier sur le web)*.
Notons qu'ici le fichier HTML et le fichier CSS sont dans le même dossier sans quoi le chemin relatif pointerait nulle part.
